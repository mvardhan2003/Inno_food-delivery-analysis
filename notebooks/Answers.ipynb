```python
def ratings(r):
    if r >= 3.0 and r <= 3.5:
        return 'a'
    elif r >= 3.6 and r <= 4.0:
        return 'b'
    elif r >= 4.1 and r <= 4.5:
        return 'c'
    elif r >= 4.6 and r <= 5.0:
        return 'd'
```

```python
df['rating_group'] = df['rating'].apply(ratings)
df['rating_group']
```

```python
df.groupby(by='rating_group')['total_amount'].sum()
```
