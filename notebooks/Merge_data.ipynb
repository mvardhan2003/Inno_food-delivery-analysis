```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("../data/orders.csv")
orders.head()
```

```python
users = pd.read_json("../data/users.json")
users.head()
```

```python
conn = sqlite3.connect("../data/restaurants.db")
cursor = conn.cursor()
```

```python
with open("../data/restaurants.sql", "r", encoding="utf-8") as f:
    sql_script = f.read()

cursor.executescript(sql_script)
conn.commit()
```

```python
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
final_df = orders.merge(users, on="user_id", how="left") \
                 .merge(restaurants, on="restaurant_id", how="left")
```

```python
final_df
```

```python
final_df.to_csv("../data/final_food_delivery_dataset.csv", index=False)
```
